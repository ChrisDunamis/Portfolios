___(5): Find the outlier___

Take a look at the scatterplot that correlates Pokémons Attack and Defense. What's the `Name` of the Pokémon that is a clear outlier with strong `Defense`, but very low `Attack`).

```python
sns.set_style(
    "ticks", 
    {"axes.grid" : True}
)

ax = sns.scatterplot(data=df, x="Defense", y="Attack")

ax.set_title("Pokémons That Are Clearly Outliers in Terms of Attack / Defense")
ax.annotate(
    "Who's this guy?", xy=(228, 10), xytext=(150, 10), color="red", arrowprops=dict(arrowstyle="->", color="black")
);
```

```python
outlier_result = df.sort_values(by=["Defense", "Attack"], ascending=[False, True])

print("From the table below, the Pokémon that is a clear outlier in terms of strong Defense, but very low Attack is, " + 
      "'{pokemon_name}'.".format(pokemon_name=outlier_result.head(1).sum()[0]
      ))
outlier_result.head()
```

## __Advance Selection__

Now let's use boolean operators to create more advanced expressions.

___(6): How many Fire-Flying Pokémons are there?___

How many Fire-Flying Pokémons are of `Type 1` Fire and `Type 2` Flying?

```python
print("Sum of Type 1 Fire Pokémons:   {type1_fire}".format(type1_fire=(df["Type 1"] == "Fire").sum()) + "\n" + 
      "Sum of Type 2 Flying Pokémons: {type2_flying}".format(type2_flying=(df["Type 2"] == "Flying").sum()) + "\n\n" + 
      
      "Sum of Type 1 Fire and Type 2 Flying Pokémons: {sum_type1_and_type2}".format(
          sum_type1_and_type2=(
              (df["Type 1"] == "Fire") & (df["Type 2"] == "Flying")
          ).sum()
      )
)
```

```python
df.loc[
    (df["Type 1"] == "Fire") & (df["Type 2"] == "Flying")
]
```

<br>

___(7): How many 'Poison' Pokémons are there?___

How many Pokémons exist that are of type `Poison` in either `Type 1` or `Type 2`?

```python
print("Sum of Type 1 Poison Pokémons: {type1_poison}".format(type1_poison=(df["Type 1"] == "Poison").sum()) + "\n" + 
      "Sum of Type 2 Poison Pokémons: {type2_poison}".format(type2_poison=(df["Type 2"] == "Poison").sum()) + "\n\n" + 
      
      "Sum of Type 1 and Type 2 Poison Pokémons: {type1_and_type2_poison}".format(
          type1_and_type2_poison=(
              (df["Type 1"] == "Poison") | (df["Type 2"] == "Poison")
          ).sum()
      )
)
```

```python
df.query("`Type 1` == 'Poison' or `Type 2` == 'Poison'")
```

___(8): What Pokémon of `Type 1` Ice has the strongest defense?___

Enter its name below...

```python
df.loc[
    (df["Type 1"] == "Ice") & (
        df["Defense"] == df.loc[
            (df["Type 1"] == "Ice", "Defense")
        ].max()
    )
].head()
```

```python
print("The name of the Pokémon that has the strongest defense with Type 1 Ice is: {pokemon!r}.".format(
        pokemon=df.loc[(df["Type 1"] == "Ice")].sort_values("Defense", ascending=False)[0:1].iloc[0, 0]
    )
)
```
